```python
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
```

```python
df_kwh = pd.read_csv('../gregorydrdata_kwh.csv',encoding='ascii',index_col='end_date')
df_kwh.index = [datetime(int(val.split('/')[2]),int(val.split('/')[0]),int(val.split('/')[1])) for val in df_kwh.index]
```

```python
plt.plot(df_kwh['kwh'].resample('12M'),'-')
plt.ylabel('kWh')
plt.title('Energy Usage at Gregory Drive')
plt.savefig('yearly.jpg',dpi=200)
```

```python
slopes = {}
intercepts = {}
for year in range(2008,2016):
    hdd_df = df_kwh[df_kwh['cdd65']>0]
    print 'Mean kWh in summer of '+str(year)+': ' + str(hdd_df[hdd_df.index.year==year]['kwh'].median())
    #regr = linear_model.LinearRegression()
    x_val = np.array(hdd_df[hdd_df.index.year==year]['cdd65']).reshape(-1,1)
    #regr.fit(x_val,hdd_df[hdd_df.index.year==year]['kwh'])
    X = hdd_df[hdd_df.index.year==year]['cdd65']
    X = sm.add_constant(X)
    model = sm.OLS(hdd_df[hdd_df.index.year==year]['kwh'],X)
    results = model.fit()
    plt.figure(1)
    plt.plot(hdd_df[hdd_df.index.year==year]['cdd65'],hdd_df[hdd_df.index.year==year]['kwh'],'.',label=str(year))
    plt.legend(bbox_to_anchor=(1.1, 1.05))
    plt.xlabel('CDD')
    plt.ylabel('kWh')
    plt.figure(2)
    plt.plot(hdd_df[hdd_df.index.year==year]['cdd65'],results.params['cdd65']*hdd_df[hdd_df.index.year==year]['cdd65']+\
             results.params['const'],label=str(year),lw=3)
    plt.legend(bbox_to_anchor=(1.1, 1.05))
    plt.xlabel('CDD')
    plt.ylabel('kWh')
    slopes[year] =results.params['cdd65']
    intercepts[year] = results.params['const']
    
    '''
    plt.figure(3)
    plt.plot(hdd_df[hdd_df.index.year==year]['cdd65'],results.params['cdd65']*hdd_df[hdd_df.index.year==year]['cdd65']\
             ,label=str(year))
    plt.legend(bbox_to_anchor=(1.1, 1.05))
    plt.xlabel('CDD')
    plt.ylabel('kWh')
    '''
print
hdd_df = df[df['cdd65']>0]
year = 2013
comparison = 2012
print 'kWh used for AC in '+str(year)+': ' + str(np.sum(hdd_df[hdd_df.index.year == year]['cdd65']*slopes[year]))
print 'kWh that would have been usd for AC with 2012 insulation '+str(np.sum(hdd_df[hdd_df.index.year == year]['cdd65']*slopes[comparison]))
print 'kWh saved in '+str(year)+': '+str(100-((np.sum(hdd_df[hdd_df.index.year == year]['cdd65']*\
                               slopes[comparison])-np.sum(hdd_df[hdd_df.index.year == year]['cdd65']*slopes[year]))/\
                                          np.sum(hdd_df[hdd_df.index.year == year]['cdd65']*slopes[comparison])*100))+'%'
print
hdd_df = df[df['cdd65']>0]
year = 2014
comparison = 2012
print 'kWh used for AC in '+str(year)+': ' + str(np.sum(hdd_df[hdd_df.index.year == year]['cdd65']*slopes[year]))
print 'kWh that would have been usd for AC with 2012 insulation '+str(np.sum(hdd_df[hdd_df.index.year == year]['cdd65']*slopes[comparison]))
print 'kWh saved in '+str(year)+': '+str(100-((np.sum(hdd_df[hdd_df.index.year == year]['cdd65']*\
                               slopes[comparison])-np.sum(hdd_df[hdd_df.index.year == year]['cdd65']*slopes[year]))/\
                                          np.sum(hdd_df[hdd_df.index.year == year]['cdd65']*slopes[comparison])*100))+'%'
plt.savefig('kwhcdd.jpg',dpi=200)
```
